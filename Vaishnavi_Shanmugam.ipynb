```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
#  dataset
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Fiona', 'George', 'Harry'],
    'Class': ['Math', 'Math', 'Science', 'Math', 'Science', 'Math', 'Science', 'Science'],
    'Test_Score': [85, 78, 92, 67, 88, 72, 90, 75]
}
df = pd.DataFrame(data)

```

#Problem 1.1: Group-By and Mean Calculation (20 points)

```python
#Problem 1.1: Group-By and Mean Calculation (20 points)
# Perform group-by operation on 'Class' and calculate mean of 'Test_Score'
mean_scores = df.groupby('Class')['Test_Score'].mean().reset_index()
mean_scores.index = mean_scores.index + 1
# result
print(tabulate(mean_scores, headers='keys', tablefmt='grid'))
```

#Problem 1.2. Visualize Score Distribution (20 points)


Bar chart

```python
#Problem 1.2. Visualize Score Distribution (20 points)
#  bar chart
plt.figure(figsize=(8, 6))
bars=plt.bar(df['Student'], df['Test_Score'], color='steelblue')

#  labels and title
plt.title('Test Score Distribution', fontsize=14)
plt.xlabel('Student', fontsize=12)
plt.ylabel('Test Score', fontsize=12)

#add score in the top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval}', va='bottom', ha='center', fontsize=10)

# show the bar chart
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Histogram chart

```python
# Define the bins for range
bins = [65, 70, 75, 80, 85, 90, 95]

# histogram
plt.figure(figsize=(8, 6))
n, bins, patches = plt.hist(df['Test_Score'], bins=bins, color='steelblue', edgecolor='black')

#  labels, title
plt.title('Test Score Distribution by Range', fontsize=14)
plt.xlabel('Test Score Range', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# frequencies
for i in range(len(patches)):
    plt.text(patches[i].get_x() + patches[i].get_width()/2, n[i], str(int(n[i])), ha='center', va='bottom', fontsize=10)

# show the histogram
plt.xticks(bins)
plt.tight_layout()
plt.show()
```

#Problem 1.3. Descriptive Statistics Summary (10 points)





```python
from statistics import mode
#  mean, median, and mode
mean_score = df['Test_Score'].mean()
median_score = df['Test_Score'].median()
mode_score = mode(df['Test_Score'])

# Create a summary DataFrame
summary_data = {
    'Statistic': ['Mean', 'Median', 'Mode'],
    'Value': [mean_score, median_score, mode_score]
}

summary_df = pd.DataFrame(summary_data).reset_index(drop=True)
summary_df.index = summary_df.index + 1

#results
print(tabulate(summary_df, headers='keys', tablefmt='grid'))
```

#Problem 2.1 Correlation between Exercise Frequency and Happiness Levels (20 points)

```python

#  dataset
data = {
    'Participant': [1, 2, 3, 4, 5, 6, 7, 8],
    'Exercise_Frequency': [5, 2, 4, 1, 3, 0, 6, 2],
    'Happiness_Score': [9, 6, 8, 5, 7, 4, 10, 5]
}

df = pd.DataFrame(data)

# Calculate the Pearson correlation between Exercise_Frequency and Happiness_Score
correlation = df['Exercise_Frequency'].corr(df['Happiness_Score'])

# result
print(f"Correlation between Exercise_Frequency and Happiness_Score: {correlation:.2f}")

if correlation > 0:
    print("The correlation is positive.")
    print("A positive correlation means that as exercise frequency increases, happiness scores tend to increase as well.")
elif correlation < 0:
    print("The correlation is negative.")
    print("A negative correlation means that as exercise frequency increases, happiness scores tend to decrease.")
else:
    print("There is no correlation.")
    print("No correlation means that exercise frequency and happiness scores do not seem to be related.")
```

#Problem 2.2 Create a scatter plot with Exercise_Frequency on the x-axis and Happiness_Score on the y-axis. Each point represents a participant's data. (20 points)


```python
plt.figure(figsize=(8, 6))

#  scatter plot
sns.regplot(x='Exercise_Frequency', y='Happiness_Score', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})

for i in range(len(df)):
    plt.text(df['Exercise_Frequency'][i] + 0.1, df['Happiness_Score'][i],
             f'({df["Exercise_Frequency"][i]}, {df["Happiness_Score"][i]})', fontsize=9)

#  labels and title
plt.title('Scatter Plot: Exercise Frequency vs Happiness Score', fontsize=14)
plt.xlabel('Exercise Frequency (Days/Week)', fontsize=12)
plt.ylabel('Happiness Score (1-10)', fontsize=12)

# show the plot
plt.tight_layout()
plt.show()
```

#Problem 2.3 calculate the average Happiness_Score and Exercise_Frequency! And also, return the best high value for Happiness_Score! (10 points)

```python
# average values
average_happiness = df['Happiness_Score'].mean()
average_exercise = df['Exercise_Frequency'].mean()

# Find the highest value for Happiness_Score
highest_happiness = df['Happiness_Score'].max()

# results
print(f"Average Happiness Score: {average_happiness:.2f}")
print(f"Average Exercise Frequency: {average_exercise:.2f}")
print(f"Highest Happiness Score: {highest_happiness}")
```

```python
labels = ['Avg Happiness Score', 'Avg Exercise Frequency', 'Highest Happiness Score']
values = [average_happiness, average_exercise, highest_happiness]

#  bar chart
plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['blue', 'green', 'red'])

#  labels and title
plt.title('Summary Statistics', fontsize=14)
plt.ylabel('Value', fontsize=12)

#  value labels
for i, v in enumerate(values):
    plt.text(i, v + 0.1, f'{v:.2f}', ha='center', fontsize=12)

# show the plot
plt.tight_layout()
plt.show()
```
